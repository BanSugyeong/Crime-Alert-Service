```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
pop = pd.read_csv('1인가구.csv', encoding='cp949')
pop
```

```python
# 1인가구 데이터 남여 구분하여 새로운 데이터프레임 생성
pop_female = pop[pop['성별(1)'] == '여자']
pop_male = pop[pop['성별(1)'] == '남자']

pop_female.rename(columns={'2020_20대':'2020_20대_여자', '2020_30대':'2020_30대_여자', '2020_40대':'2020_40대_여자', '2020_50대':'2020_50대_여자', 
                           '2021_20대':'2021_20대_여자', '2021_30대':'2021_30대_여자', '2021_40대':'2021_40대_여자', '2021_50대':'2021_50대_여자', 
                           '2022_20대':'2022_20대_여자', '2022_30대':'2022_30대_여자', '2022_40대':'2022_40대_여자', '2022_50대':'2022_50대_여자'}, inplace=True)
pop_male.rename(columns={'2020_20대':'2020_20대_남자', '2020_30대':'2020_30대_남자', '2020_40대':'2020_40대_남자', '2020_50대':'2020_50대_남자', 
                         '2021_20대':'2021_20대_남자', '2021_30대':'2021_30대_남자', '2021_40대':'2021_40대_남자', '2021_50대':'2021_50대_남자', 
                         '2022_20대':'2022_20대_남자', '2022_30대':'2022_30대_남자', '2022_40대':'2022_40대_남자', '2022_50대':'2022_50대_남자'}, inplace=True)


del pop_female['성별(1)']
del pop_male['성별(1)']

pop_female.to_csv('1인가구_여자.csv', encoding='cp949', index=False)
pop_male.to_csv('1인가구_남자.csv', encoding='cp949', index=False)
```
